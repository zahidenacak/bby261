```python
import os

DOSYA_ADI = "kutuphane.txt"

def dosya_oku_ve_kitaplari_yukle():
    kitaplar_listesi = []
    if os.path.exists(DOSYA_ADI):
        with open(DOSYA_ADI, 'r', encoding='utf-8') as dosya:
            for satir in dosya:
                parcalar = satir.strip().split(',')
                if len(parcalar) == 2:
                    ad = parcalar[0]
                    yazar = parcalar[1]
                    kitap = {'ad': ad, 'yazar': yazar}
                    kitaplar_listesi.append(kitap)
    return kitaplar_listesi

def verileri_dosyaya_yaz(kitaplar_listesi):
    with open(DOSYA_ADI, 'w', encoding='utf-8') as dosya:
        for kitap in kitaplar_listesi:
            dosya.write(f"{kitap['ad']},{kitap['yazar']}\n")


# --- ANA PROGRAM BAŞLANGICI ---

kitaplar = dosya_oku_ve_kitaplari_yukle()
print("Kütüphane programına hoş geldiniz! Mevcut kayıtlar yüklendi.")

while True:
    print("\n--- KÜTÜPHANE MENÜSÜ ---")
    print("1: Kitap Ekle")
    print("2: Tüm Kitapları Listele")
    print("3: Kitap Ara")
    print("4: Çıkış")

    secim = input("Yapmak istediğiniz işlemi seçin: ").strip()

    if secim == '1':
        print("\n-> Yeni Kitap Ekleme")
        ad = input("Kitabın Adı: ")
        yazar = input("Yazarın Adı: ")
        yeni_kitap = {'ad': ad, 'yazar': yazar}
        kitaplar.append(yeni_kitap)

        verileri_dosyaya_yaz(kitaplar)
        print(f"'{ad}' adlı kitap başarıyla kaydedildi!")

    elif secim == '2':
        print("\n-> Kütüphanedeki Kitaplar")
        if not kitaplar:
            print("Kütüphanenizde henüz kitap yok.")
        else:
            for i, kitap in enumerate(kitaplar, start=1):
                print(f"{i}. Kitap Adı: {kitap['ad']}, Yazar: {kitap['yazar']}")

    elif secim == '3':
        print("\n-> Kitap Arama")
        if not kitaplar:
            print("Arama yapmak için kütüphanede en az bir kitap olmalı.")
        else:
            aranan_kelime = input("Aramak istediğiniz kelimeyi girin: ").lower()
            bulunanlar = []
            for kitap in kitaplar:
                if aranan_kelime in kitap['ad'].lower() or aranan_kelime in kitap['yazar'].lower():
                    bulunanlar.append(kitap)

            if not bulunanlar:
                print("Aradığınız kriterlere uygun kitap bulunamadı.")
            else:
                print("\n--- Arama Sonuçları ---")
                for i, bulunan_kitap in enumerate(bulunanlar, start=1):
                    print(f"{i}. Kitap Adı: {bulunan_kitap['ad']}, Yazar: {bulunan_kitap['yazar']}")

    elif secim == '4':
        print("Programdan çıkılıyor. Hoşça kalın!")
        break

    else:
        print("Hatalı seçim! Lütfen 1, 2, 3 veya 4 seçeneklerinden birini girin.")
```
